```python
from collections import deque


def W(d1, d2):
    result = deque()
    
    while len(d1) > 0 and len(d2) > 0:
        result.append(d1.popleft())
        result.append(d2.popleft())
    
    return result


deq1 = deque('qwertyuiop')
deq2 = deque('123456')
print(W(deq1, deq2))
```

```python
def genfromq(d1, d2):
    while True:
        try:
            k1 = d1.popleft()
            k2 = d2.popleft()
            yield k1
            yield k2
        except:
            break

            
deq1 = deque('qwertyuiop')
deq2 = deque('123456')

gen = genfromq(deq1, deq2)
# print(list(gen))
```
