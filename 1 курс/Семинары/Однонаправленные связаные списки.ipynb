#### Создать класс однонаправленных связных списков. Осуществить добавление элементов слева, справа, между элементами (узлами) списка.

```python
class Node:
    
    def __init__(self, value=None, nxt=None):
        self.obj = value
        self.value = value
        self.nxt = nxt
        
    def __str__(self):
        return f'{self.value}'
    
    def __gt__(self, other):
        return self.value > other.value
    
    def __le__(self, other):
        return (self.value < other.value) or (self.value == other.value)
```

```python
class SingleLinkedList:
    
    def __init__(self):
        self.head = None
        self.tail = None
    
    def append(self, obj):
        new_obj = Node(obj)
        if self.is_empty():
            self.head = new_obj
            self.tail = new_obj
        elif self.head == self.tail:
            self.tail = new_obj
            self.head.nxt = self.tail
        else:
            self.tail.nxt = new_obj
            self.tail = new_obj
    
    def append_left(self, obj):
        new_obj = Node(obj)
        if self.is_empty():
            self.head = new_obj
            self.tail = new_obj
        elif self.head == self.tail:
            self.head = new_obj
            self.head.nxt = self.tail
        else:
            a = self.head
            self.head = new_obj
            self.head.nxt = a
        
    def append_sr(self, obj, k):
        new_obj = Node(obj)
        if self.head != self.tail and self.is_empty() == False:
            current = self.head
            while current.nxt is not None:
                d = current.value
                if d==k:
                    h = current.nxt
                    current.nxt = new_obj
                    new_obj.nxt=h
                current = current.nxt
    
    def __str__(self):
        if not self.is_empty():
            current = self.head
        
            out = f'SingleLinkedList[ {current} '
            while current.nxt is not None:
                current = current.nxt
                out += str(current) + ' '
            return out + ']'
        else:
            return 'SingleLinkedList[]'
    
    def is_empty(self):
        return self.head is None

    def printf(self):
        current = self.head
        while current.nxt is not None:
            print(current, end = ' ')
            current = current.nxt
            
    # 1. Написать программу передвижения элемента на n позиций.
    def move(self, elem, n):
        if not self.is_empty():
            current = self.head
            if current.value == elem:
                    el == current
            else:
                prev = current
                while current.nxt is not None:
                    current = current.nxt
                    if current.value == elem:
                        el = current
                        prev.nxt = el.nxt
                        break
                    prev = current
                else:
                    raise Exception("Нет такого элемента")
            for i in range(n):
                if current.nxt is not None:
                    current = current.nxt
                else:
                    raise Exception("Превышена длина списка")
            el.nxt = current.nxt
            current.nxt = el
            
    # 2. Вставить элемент после n-го элемента списка.
    def insert(self, elem, n):
        if not self.is_empty():
            current = self.head
            for i in range(n-1):
                if current.nxt is not None:
                    current = current.nxt
                else:
                    raise Exception("Превышена длина списка")
            new_el = Node(elem)
            new_el.nxt = current.nxt
            current.nxt = new_el
    
    # 4. Соединить два списка.
    def extend(self, lst):
        if self.head is not None:
            self.tail.nxt = lst.head
        else:
            self.head = lst.head
    
    # 5. Удалить n-ый элемент из списка.
    def delete(self, n):
        if not self.is_empty():
            current = self.head
            if n == 0:
                self.head = current.nxt
                current.nxt = None
            else:
                for i in range(n-1):
                    if current.nxt is not None:
                        current = current.nxt
                    else:
                        raise Exception("Превышена длина списка")
                current.nxt = current.nxt.nxt
        
    # 6. Создать копию списка.
    def copy(self):
        new_list = SingleLinkedList()
        if self.head is None:
            return new_list
        else:
            current = self.head
            new_node = Node(current.value)
            new_list.append(new_node)
        while current.nxt is not None:
            current = current.nxt
            new_node = Node(current.value)
            new_list.append(new_node)
        return new_list
    
    # 7. Создать список, содержащий элементы общие для двух списков.
    @staticmethod
    def common_el(list1, list2):
        def search(lst, el):
            current = lst.head
            if current.value == el:
                return True
            else:
                while current.nxt is not None:
                    current = current.nxt
                    if current.value == el:
                        return True
                return False
        new_list = SingleLinkedList()
        current = list1.head
        if search(list2, current.value):
            new_list.append(current.value)
        while current.nxt is not None:
            current = current.nxt
            if search(list2, current.value):
                new_list.append(current.value)
        return new_list
    
    # 8. Упорядочить элементы списка по возрастанию.
    def sort(self):
        if not self.is_empty():
            current = self.head
            nxt = current.nxt
            c2 = nxt
#             print('c2:', c2)
            c3 = self.head
            if c3.__gt__(c2):
                nxt = c2.nxt
                c2.nxt = c3
                self.head = c2
                c3.nxt = nxt
#             print(self)
            while nxt is not None:
                c2 = nxt
                c3 = self.head
                prev = None
#                 print('c2:', c2)
                while c3 is not nxt:
#                     print('nxt:', nxt)
#                     print('c3:', c3)
                    if c3.__gt__(c2):
#                         print('prev', prev)
                        nxt = c2.nxt
                        c2.nxt = c3
                        if prev == None:
                            self.head = c2
                        else:
                            prev.nxt = c2
                        while c3.nxt is not c2:
                            c3 = c3.nxt
                        c3.nxt = nxt
                        break
                    prev = c3
                    c3 = c3.nxt
                c3.nxt = nxt
            self.tail = c3
#                 self.printf()

    # 9. Удалить каждый второй элемент списка.
    def del_even(self):
        current = self.head
        self.head = current.nxt
        current = self.head
        while current is not None:
            prev = current
            current = current.nxt
            if current:
                nxt = current.nxt
                prev.nxt = nxt
                current = nxt
                
    # 10. Удалить каждый третий элемент списка.
    def del_every_third(self):
        current = self.head
        self.head = current.nxt
        current = self.head
        while current is not None:
            current = current.nxt
            prev = current
            current = current.nxt
            if current:
                nxt = current.nxt
                prev.nxt = nxt
                current = nxt
    
    # 11. Упорядочить элементы списка по убыванию.
    def sort_rev(self):
        self.sort()
        self.head, self.tail = self.tail, self.head
        prev = self.tail
        current = prev.nxt
        nxt = current.nxt
        while current is not self.head:
            current.nxt = prev
            prev = current
            current = nxt
            nxt = current.nxt
        self.tail.nxt = None
        current.nxt = prev
    
    # 12. Очистить список.
    def clear(self):
        self.tail, self.head = None, None
```
